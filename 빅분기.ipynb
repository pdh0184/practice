Q57. Income_Category의 카테고리를 map 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라
- Unknown : N

- Less than $40K : a

- $40K - $60K : b

- $60K - $80K : c

- $80K - $120K : d

- $120K +’ : e

```python
df["new_Income"] = df['Income_Category'].map({
    "Unknown" : "N",
    "Less than $40K" : "a",
    "$40K - $60K" : "b",
    "$60K - $80K" : "c",
    "$80K - $120K" : "d",
    "$120K +’" : "e"})
df
```

Q58. Income_Category의 카테고리를 apply 함수를 이용하여 다음과 같이 변경하여 newIncome 컬럼에 매핑하라

```python
def changeCategory(x):
    if x =='Unknown':
        return 'N'
    elif x =='Less than $40K':
        return 'a'
    elif x =='$40K - $60K':
        return 'b'
    elif x =='$60K - $80K':
        return 'c'
    elif x =='$80K - $120K':
        return 'd'
    elif x =='$120K +' :
        return 'e'

df['newIncome']  =df.Income_Category.apply(changeCategory)

Ans = df['newIncome']
Ans
```
